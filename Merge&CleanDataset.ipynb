```python
# import required module
import os
# assign directory
super_directory = '/content/drive/MyDrive/AI_Industry/FlightsDataset/'
dfs = []
 
# iterate over files in that directory
for directory in os.listdir(super_directory):
    directory_path = os.path.join(super_directory, directory)
    for file in os.listdir(directory_path):
        f = os.path.join(directory_path, file)
        # checking if it is a file
        if os.path.isfile(f):
            dfs.append(pd.read_csv(f, encoding = "ISO-8859-1"))

df = pd.concat(dfs)
```

```python
df.head(10)
```

```python
df = df.iloc[: , 1:]
df.head(10)
df = df[df["Period"] != "Two Months"]
df = df[df["Period"] != "Three Months"]
df = df[df["Period"] != "Four Months"]
```

```python
# take only one flight per day (the one with the minimum price)
df = df.groupby(["Origin", "Destination", "Date", "Tracking Date"], as_index=False)["Price"].min()
df = df.reset_index()

for index, row in df.iterrows():
    df.loc[index, "Date"] = row["Date"].replace(" ", "")
df = df.iloc[:, 1:]
```

```python
df.to_csv('/content/drive/MyDrive/AI_Industry/dataset.csv')
```
